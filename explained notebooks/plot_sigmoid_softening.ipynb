```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x, softening_coeff):
    return 2 * (1 / (1 + np.exp(-x / softening_coeff)) - 0.5)

values = np.arange(0, 10, 0.1)
plt.figure()  # create a new figure

for softening_coeff, color in zip(range(1, 5), ['r', 'g', 'b', 'm']):  # assigning colors for each plot
    plt.plot(values, sigmoid(values, softening_coeff), label=f'Softening Coeff: {softening_coeff}', color=color)

plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Softening Comparison')
plt.legend()  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x, softening_coeff):
    return 2 * (1 / (1 + np.exp(-x / softening_coeff)) - 0.5)

values = np.arange(-10, 10, 0.1)
plt.figure()  # create a new figure

for softening_coeff, color in zip(range(1, 5), ['r', 'g', 'b', 'm']):  # assigning colors for each plot
    plt.plot(values, sigmoid(values, softening_coeff), label=f'Softening Coeff: {softening_coeff}', color=color)

plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Softening Comparison with negatives')
#we can see that the last one is more linear and therefore more compatible with negotiation paradigm. 
#coeff=1 is the original sigmoid. The others are softened.
plt.legend()  
plt.show()
```
